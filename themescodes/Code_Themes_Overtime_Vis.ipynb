```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## General Data Stuff
Loading in the data files, modifying dataframes, straightforward plotting

```python
df = pd.read_csv('/Users/rachelroca/Documents/GitHub/DAAAAMNdjmm/data cleaning/codes-by-time-bucket.csv')
theme_df= pd.read_csv("/Users/rachelroca/Documents/GitHub/DAAAAMNdjmm/codes_hypergraph/Code Book.csv")
```

```python
#Absolute
for i in range(df.shape[0]):
    plt.plot(['preJMM 20', 'JMM 20', 'InterJMM', 'JMM 21', 'Post JMM'], df.iloc[i,1:], '*-', label=df.iloc[i,0])
```

```python
themes_list = []
for i in range(df.shape[0]):
    themes_list.append(np.array(theme_df.columns[theme_df.isin([df.iloc[i,0]]).any()])[0])
```

```python
df['themes']=themes_list
```

## Relative 
(denominator = sum of codes for each time slice), also I like percents

```python
df_rel = df.copy()
df_rel["Pre JMM 2020"] = df_rel["Pre JMM 2020"]/df_rel["Pre JMM 2020"].sum() * 100
df_rel["JMM 2020"] = df_rel["JMM 2020"]/df_rel["JMM 2020"].sum() * 100
df_rel["Inter JMM"] = df_rel["Inter JMM"]/df_rel["Inter JMM"].sum() * 100
df_rel["JMM 2021"] = df_rel["JMM 2021"]/df_rel["JMM 2021"].sum() * 100
df_rel["Post JMMs"] = df_rel["Post JMMs"]/df_rel["Post JMMs"].sum() *100
df_rel
```

```python
#Relative
for i in range(df_rel.shape[0]):
    plt.plot(['preJMM 20', 'JMM 20', 'InterJMM', 'JMM 21', 'Post JMM'], df_rel.iloc[i,1:-1], '*-', label=df_rel.iloc[i,0])
```

```python

df_rel_by_theme = df_rel.groupby(['themes']).sum()
for i in range(df_rel_by_theme.shape[0]):
    plt.plot(['Pre JMM 20', 'JMM 20', 'InterJMM', 'JMM 21', 'Post JMM'], df_rel_by_theme.iloc[i,:], '*-',)
plt.legend(['SJEDI', 'BCP', 'BC', 'CCM', 'SO'])
```
