```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/Users/rachelroca/Documents/GitHub/DAAAAMNdjmm/data cleaning/codes-by-time-bucket.csv')
theme_df= pd.read_csv("/Users/rachelroca/Documents/GitHub/DAAAAMNdjmm/codes_hypergraph/Code Book.csv")
```

```python
for i in range(df.shape[0]):
    plt.plot(['preJMM 20', 'JMM 20', 'InterJMM', 'JMM 21', 'Post JMM'], df.iloc[i,1:], '*-', label=df.iloc[i,0])
```

```python
themes_list = []
for i in range(df.shape[0]):
    themes_list.append(np.array(theme_df.columns[theme_df.isin([df.iloc[i,0]]).any()])[0])
```

```python
df['themes']=themes_list
```

```python
df_by_theme = df.groupby(['themes']).sum()
for i in range(df_by_theme.shape[0]):
    plt.plot(['Pre JMM 20', 'JMM 20', 'InterJMM', 'JMM 21', 'Post JMM'], df_by_theme.iloc[i,:], '*-',)
plt.legend(['SJEDI', 'BCP', 'BC', 'CCM', 'SO'])
```
